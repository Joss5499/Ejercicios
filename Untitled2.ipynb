```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4]
y = [1, 2, 0, 0.5]

# Crear figura y ejes
fig, ax = plt.subplots()

# Dibujar los puntos (gráfico de dispersión)
ax.scatter(x, y, color='red', marker='o', label='Datos')

# Añadir título y etiquetas
ax.set_title('Gráfico de Puntos')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.legend()

# Guardar la figura como imagen PNG
fig.savefig('grafico_puntos.png', dpi=300)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos para la "Z"
x = [0, 1, 2, 0]  # Coordenadas en el eje X
y = [0, 0, 2, 2]  # Coordenadas en el eje Y

# Crear figura y ejes
fig, ax = plt.subplots()

# Dibujar la Z usando plot (puntos rojos y líneas conectadas)
ax.plot(x, y, 'ro-', label='Forma de Z')  # 'ro-' para puntos rojos y líneas conectadas

# Añadir título y etiquetas
ax.set_title('Gráfico de una Z perfecta')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')

ax.set_xticks(range(-1, 4))  # Ajustar los ticks del eje X
ax.set_yticks(range(-1, 4))  # Ajustar los ticks del eje Y

ax.legend()

fig.savefig('grafico_z_perfecta.png', dpi=300)

plt.show()


```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4]
y = [2, 3, 0, 0.5]

# Crear figura y ejes
fig, ax = plt.subplots()

# Dibujar el diagrama de áreas
ax.fill_between(x, y, color="skyblue", alpha=0.5, label="Área bajo la curva")

# Añadir título y etiquetas
ax.set_title('Diagrama de Áreas')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')

# Añadir leyenda
ax.legend()

# Guardar la figura como imagen PNG
fig.savefig('diagrama_areas.png', dpi=300)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

edades = []

for i in range(1, 6):
    edad = int(input(f"Introduce la edad del alumno {i}: "))
    edades.append(edad)

media = sum(edades) / len(edades)

# Crear etiquetas de los alumnos para el gráfico
alumnos = [f"Alumno {i}" for i in range(1, 6)]

fig, ax = plt.subplots()

# Crear el diagrama de barras verticales con las edades
ax.bar(alumnos, edades, color='skyblue', edgecolor='black', label='Edades')

# Añadir una línea horizontal para la media
ax.axhline(y=media, color='red', linestyle='--', label=f'Media = {media:.2f}')

ax.set_title('Edades de los Alumnos')
ax.set_xlabel('Alumnos')
ax.set_ylabel('Edades')

ax.legend()

fig.savefig('edades_alumnos.png', dpi=300)

plt.show()

print(f"\nLa media de las edades es: {media:.2f}")
```

```python
import matplotlib.pyplot as plt


edades = []


for i in range(1, 6):
    edad = int(input(f"Introduce la edad del alumno {i}: "))
    edades.append(edad)

media = sum(edades) / len(edades)

promedio = sum(edades) / len(edades)

# Crear etiquetas de los alumnos para el gráfico
alumnos = [f"Alumno {i}" for i in range(1, 6)]

# Crear figura y ejes para los subgráficos (dos gráficos en una fila)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# **Primer gráfico - Media**
ax1.bar(alumnos, edades, color='skyblue', edgecolor='black', label='Edades')
ax1.axhline(y=media, color='red', linestyle='--', label=f'Media = {media:.2f}')
ax1.set_title('Gráfico de Edades con Media')
ax1.set_xlabel('Alumnos')
ax1.set_ylabel('Edades')
ax1.legend()

# **Segundo gráfico - Promedio**
ax2.bar(alumnos, edades, color='lightgreen', edgecolor='black', label='Edades')
ax2.axhline(y=promedio, color='green', linestyle='--', label=f'Promedio = {promedio:.2f}')
ax2.set_title('Gráfico de Edades con Promedio')
ax2.set_xlabel('Alumnos')
ax2.set_ylabel('Edades')
ax2.legend()

# Ajustar el layout para evitar solapamientos
plt.tight_layout()

fig.savefig('edades_media_promedio.png', dpi=300)

plt.show()

print(f"\nLa media de las edades es: {media:.2f}")
print(f"El promedio de las edades es: {promedio:.2f}")
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo (número de votos para cada candidato)
votos = [30, 40, 20, 10, 50]

# Etiquetas para cada sector (candidatos)
etiquetas = ['Candidato A', 'Candidato B', 'Candidato C', 'Candidato D', 'Candidato E']

# Crear el diagrama de sectores (gráfico circular)
plt.pie(votos, labels=etiquetas, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange', 'pink', 'lightcoral'])

# Añadir título
plt.title('Distribución de Votos por Candidato')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

pesos = []
for i in range(1, 6):
    peso = float(input(f"Introduce el peso del estudiante {i} (en kg): "))
    pesos.append(peso)

media = np.mean(pesos)
desviacion_estandar = np.std(pesos)

# Datos para los gráficos (los valores representan las partes de la media y desviación estándar en el total)
total = sum(pesos)

# Crear etiquetas para los gráficos
etiquetas_media = ['Media', 'Resto']  # Media vs. el resto
etiquetas_desviacion = ['Desviación Estándar', 'Resto']  # Desviación Estándar vs. el resto

# Gráfico de la Media
plt.figure(figsize=(12, 6))  # Ajustamos el tamaño de la figura
plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, gráfico 1
plt.pie([media, total - media], labels=etiquetas_media, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgrey'])
plt.title('Distribución de la Media de los Pesos')

# Gráfico de la Desviación Estándar
plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, gráfico 2
plt.pie([desviacion_estandar, total - desviacion_estandar], labels=etiquetas_desviacion, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightgrey'])
plt.title('Distribución de la Desviación Estándar de los Pesos')

plt.tight_layout()
plt.show()

print(f"\nLa media de los pesos es: {media:.2f} kg")
print(f"La desviación estándar de los pesos es: {desviacion_estandar:.2f} kg")
```

```python
import matplotlib.pyplot as plt

varones = int(input("Introduce el número de estudiantes varones: "))
mujeres = int(input("Introduce el número de estudiantes mujeres: "))

# Calcular el total de estudiantes
total_estudiantes = varones + mujeres

# Calcular el porcentaje de varones y mujeres
porcentaje_varones = (varones / total_estudiantes) * 100
porcentaje_mujeres = (mujeres / total_estudiantes) * 100

etiquetas = ['Varones', 'Mujeres']

# Crear el diagrama de sectores
plt.figure(figsize=(7, 7))  # Tamaño de la figura
plt.pie([porcentaje_varones, porcentaje_mujeres], labels=etiquetas, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'pink'])

plt.title('Distribución de Estudiantes por Género')

plt.show()

print(f"\nEl porcentaje de estudiantes varones es: {porcentaje_varones:.2f}%")
print(f"El porcentaje de estudiantes mujeres es: {porcentaje_mujeres:.2f}%")
```
