```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load the dataset into a Pandas DataFrame
df = pd.read_csv("HistoricalPrices.csv")

# Rename the column to remove an additional space
df = df.rename(columns = {' Open': 'Open', ' High': 'High', ' Low': 'Low', ' Close': 'Close'})

# Convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Sort the dataset in the ascending order of date
df = df.sort_values(by = 'Date')
```

```python
# Extract the date and close price columns
dates = df['Date']
closing_price = df['Close']

# Create a line plot
plt.plot(dates, closing_price)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Aos desde 2019 hasta 2025 con fracciones de ao para crear ms puntos (como cardiograma)
years = np.array([
    2019, 2019.2, 2019.4, 2019.6, 2019.8,
    2020, 2020.2, 2020.4, 2020.6, 2020.8,
    2021, 2021.2, 2021.4, 2021.6, 2021.8,
    2022, 2022.2, 2022.4, 2022.6, 2022.8,
    2023, 2023.2, 2023.4, 2023.6, 2023.8,
    2024, 2024.2, 2024.4, 2024.6, 2024.8,
    2025
])

# Ventas con picos y cadas rpidas simuladas
sales = np.array([
    120, 110, 125, 115, 130,  # 2019 cardiograma alto
     90, 70, 60, 50, 40,      # 2020 cada fuerte
     45, 55, 40, 50, 35,      # 2021 inestable
     60, 65, 55, 70, 60,      # 2022 subida
     55, 45, 65, 50, 70,      # 2023 cardiograma inestable
     60, 80, 65, 85, 75,      # 2024 recuperacin
     90                      # 2025
])

# Crear la grfica
plt.figure(figsize=(12, 6))
plt.plot(years, sales, color='darkgreen', linewidth=2, label='Ventas (Estilo Cardiograma)')
plt.title('Simulacin de ventas con patrn tipo cardiograma (20192025)', fontsize=14)
plt.xlabel('Ao', fontsize=12)
plt.ylabel('Ventas (en millones)', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter1d  # Para suavizar la curva

# Aos desde 2019 a 2025 (con fracciones para simular el cardiograma)
years = np.array([
    2019, 2019.2, 2019.4, 2019.6, 2019.8,
    2020, 2020.2, 2020.4, 2020.6, 2020.8,
    2021, 2021.2, 2021.4, 2021.6, 2021.8,
    2022, 2022.2, 2022.4, 2022.6, 2022.8,
    2023, 2023.2, 2023.4, 2023.6, 2023.8,
    2024, 2024.2, 2024.4, 2024.6, 2024.8,
    2025
])

# Ventas con picos y cadas (tipo cardiograma)
sales = np.array([
    120, 110, 125, 115, 130,   # 2019
     90, 70, 60, 50, 40,       # 2020
     45, 55, 40, 50, 35,       # 2021
     60, 65, 55, 70, 60,       # 2022
     55, 45, 65, 50, 70,       # 2023
     60, 80, 65, 85, 75,       # 2024
     90                        # 2025
])

# Suavizamos la curva con un filtro gaussiano
smoothed_sales = gaussian_filter1d(sales, sigma=2)

# Crear la grfica
plt.figure(figsize=(12, 6))

# Lnea cardiograma (picos y valles)
plt.plot(years, sales, color='darkred', linewidth=2, label='Ventas')

# Lnea suavizada (tendencia general)
plt.plot(years, smoothed_sales, color='blue', linewidth=2.5, linestyle='--', label='Tendencia Suavizada')

# Ttulos y etiquetas
plt.title('Simulacin de ventas con lnea cardiograma y curva de tendencia (20192025)', fontsize=14)
plt.xlabel('Ao', fontsize=12)
plt.ylabel('Ventas (en millones)', fontsize=12)

# Esttica
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
